```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("partikas_produktu_fotografesana_darzeni-768x512.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#   HSV (    )
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

#    (   HSV)
lower_orange = np.array([10, 150, 150])
upper_orange = np.array([18, 255, 255])

#   
mask = cv2.inRange(hsv, lower_orange, upper_orange)

#    (  )
carrot_highlighted = cv2.bitwise_and(image_rgb, image_rgb, mask=mask)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title(" ")
plt.imshow(mask, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title(" ")
plt.imshow(carrot_highlighted)
plt.axis("off")

plt.tight_layout()
plt.show()
```
