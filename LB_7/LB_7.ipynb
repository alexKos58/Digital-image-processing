```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Загружаем изображения
scene_path = 'image.jpg'
template_path = 'template.jpg'

scene = cv2.imread(scene_path)
template = cv2.imread(template_path)

scene_gray = cv2.cvtColor(scene, cv2.COLOR_BGR2GRAY)
template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

w, h = template_gray.shape[::-1]

res = cv2.matchTemplate(scene_gray, template_gray, cv2.TM_CCOEFF_NORMED)

threshold = 0.8
loc = np.where(res >= threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(scene, pt, (pt[0] + w, pt[1] + h), (0, 0, 255), 2)

plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(scene, cv2.COLOR_BGR2RGB))
plt.title('Результат: найден шаблон')
plt.axis('off')
plt.show()
```
