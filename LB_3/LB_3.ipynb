```python
from PIL import Image, ImageFilter, ImageEnhance
import matplotlib.pyplot as plt

image = Image.open("jZU1hT8Kmy0.jpg")

# Шаг 1: Удаление шума с помощью медианного фильтра
denoised_image = image.filter(ImageFilter.MedianFilter(size=3))

# Шаг 2: Повышение резкости
sharpened_image = denoised_image.filter(ImageFilter.UnsharpMask(radius=2, percent=150, threshold=3))

# Шаг 3: Повышение контраста
enhancer = ImageEnhance.Contrast(sharpened_image)
contrast_enhanced = enhancer.enhance(1.5)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Оригинальное изображение")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(contrast_enhanced)
plt.title("Улучшенное изображение")
plt.axis("off")

plt.tight_layout()
plt.show()

# Сохранение результата
contrast_enhanced.save("improve_image.jpg")
```
